```python
import numpy as np
#t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
pl.title("voltage vs time")
pl.plot(t,v,"g")
pl.grid()
pl.xlabel("time")
pl.ylabel("voltage")
pl.savefig("images/voltage-Voltage.png")
pl.show()
```

# Plot the income of arkansas from 2005 to 2013 from income.csv file

```python

import pandas as pd
df=pd.read_csv("data files/income.csv")
years=df.columns[2:]
incomeArkansas=df.values[3,2::] #arkanas index is 4 and we want only income that starts from 2 colums
pl.title("Income of Arkansas 2005-2013")
pl.plot(years,incomeArkansas)
pl.xlabel("years")
pl.ylabel("income")
pl.savefig("images/arkansasincome.png")
pl.show()
print(years)
print(incomeArkansas)

```
